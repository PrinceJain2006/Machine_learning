```python
## independent features and dependent features 
x = dataset
y = target
```

```python
### Train test split 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train , y_test = train_test_split(
    x,y,test_size = 0.30 ,random_state = 42)              ###  train for model train
                                                           #   test for model test 
```

```python
x_train
```

```python
y_test
```

```python
## import linear regression
```

```python
## standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit_transform(x_train)          #####
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
## for inversing the value
scaler.inverse_transform(x_train)
```

```python
## now import linear regression 

from sklearn.linear_model import LinearRegression

## cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()  ## create an regression object 
regression.fit(x_train,y_train)  
```

```python
### mean score error (mse)
```

```python
mse=cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=15)  # cross verifiction = 5 times
```

```python
np.mean(mse)    # mse is a truth or predected value of difference  kam se kam aana chiye   
```

```python
#### Prediction 
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
# truth value = y_test
import seaborn as sns 
sns.displot(reg_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score 
```

```python
score = r2_score(reg_pred,y_test)
```
