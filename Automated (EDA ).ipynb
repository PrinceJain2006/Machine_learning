```python
import seaborn as sns 
```

```python
df =sns.load_dataset('diamonds')    ## iris , diamond , titanic , tips 
df.head()
```

```python
df.to_csv('house_prices.csv')
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
```

```python
AV = AutoViz_Class()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
filename = "house_prices.csv"
dft = AV.AutoViz(
    filename 
)
```
