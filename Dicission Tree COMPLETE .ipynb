üîç -  Key Differences Between Forecasting and Decision Tree Prediction.

Feature	Forecasting	Decision Trees. 

üìò Purpose	Predict future values over time	Classify or regress based on input features . 

‚è±Ô∏è Data Type	Time series data (e.g., sales every month)	Structured/tabular data (e.g., age, income, etc.) .

üß† Goal	Predict what will happen next based on past trends	Predict what class or value given current conditions .

üî¢ Output	Continuous values over time (e.g., ‚Çπ sales next week)	Class label (e.g., fraud / not fraud) or regression value .

üìà Example	Forecasting stock price, weather, demand	Predicting loan approval, disease, fraud .

üßÆ Methods	ARIMA, LSTM, Exponential Smoothing, Prophet	Decision Tree, Random Forest, XGBoost .

‚úÖ Examples
‚úÖ Forecasting example:
"Based on sales for the past 12 months, what will sales be next month?"

Uses time as a key feature.

‚úÖ Decision Tree example:
"Based on amount, age, and login behavior, is this transaction fraud?:

Use decision trees on time-series features (e.g., last 7 days of sales to predict fraud)

Combine forecasting for trend + decision trees for behavior classification
üß† Conclusion:
Forecasting ‚Üí Good for time-based future predictions

Decision Trees ‚Üí Good for classification/regression using conditions

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
## indipendent features and dependent features 
x = df.iloc[:,:-1]
y = iris.target
```

```python
y
```

```python
x
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y,test_size = 0.32, random_state = 42
)
```

```python
x_train    #>::: help of train we classify dicission tree
```

```python
x_test    #>: check acurasy 
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## PostPrunning 
treemodel=DecisionTreeClassifier(max_depth=3)
```

```python
treemodel.fit(x_train,y_train)
```

```python
## construct tree 
from sklearn import tree 
plt.figure(figsize=(10,10))
tree.plot_tree(treemodel,filled =True)
```

```python
plt.show()
```

- this is the perfect example of POSTPRUNNING
-  2ND LEVEL max_depth
-  we use hiperperameter (max_depth =2 )

```python
#  35 = versicolor       
#  31 = Virginica
```

```python
## Prediction 
y_pred =treemodel.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score ,classification_report 
```

```python
score= accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
